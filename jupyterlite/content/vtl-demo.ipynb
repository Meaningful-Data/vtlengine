# VTL Engine in the browser

This notebook runs the [VTL Engine](https://github.com/Meaningful-Data/vtlengine) **entirely client-side** on the JupyterLite Pyodide kernel — no server and no install step. The first `import vtlengine` auto-loads the WebAssembly parser and the DuckDB execution engine (a few seconds), then everything runs locally in your browser.

```python
import pandas as pd
from vtlengine import run

# A VTL transformation: multiply every measure of DS_1 by 10
script = "DS_r <- DS_1 * 10;"
data_structures = {"datasets": [{"name": "DS_1", "DataStructure": [
    {"name": "Id_1", "type": "Integer", "role": "Identifier", "nullable": False},
    {"name": "Me_1", "type": "Number", "role": "Measure", "nullable": True}]}]}
datapoints = {"DS_1": pd.DataFrame({"Id_1": [1, 2, 3], "Me_1": [10, 20, 30]})}

run(script=script, data_structures=data_structures, datapoints=datapoints)["DS_r"].data
```

```python
# Validate and infer the output structure without executing (no data needed)
from vtlengine import semantic_analysis

semantic_analysis(script=script, data_structures=data_structures)
```
